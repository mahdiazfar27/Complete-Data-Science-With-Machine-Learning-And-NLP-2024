## Operators

```python
a=20
b=2

res = a+b
print(res)

res = a-b
print(res)

res = a*b
print(res)

res = a/b
print(res)

#floor division result
res = a//b
print(res)

res = a%b
print(res)

#exponential operator
res = a**b
print(res)
```
