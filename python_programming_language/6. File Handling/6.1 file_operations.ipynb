### File Operatin - Read and Write files

file handling is a crucial part of any pl. python provides built-in functions and methods to read from and write to files, both text and binary.


```python
### read a whole file

with open('example.txt', 'r') as file:
    content = file.read() # read the entire content of the file
    print(content)
```

```python
## read a file line by line
with open('example.txt', 'r') as file:
    for line in file:
        print(line)
```

```python
with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip()) # this .strip() removes the newline character
```

```python
### writing a file (overwiting)
with open('example.txt', 'w') as file: # this 'w' mode removes the previous writings and add only the new one.
    # so its dangerous
    file.write("this is newly added line by file.write() in 'w' mode.\n")
```

```python
### writing a file (without overwriting)
with open('example.txt', 'a') as file:
    file.write("This is also a newly added line \nbut without overwriting the previous one using 'a' append mode")
```

```python
### writing list of lines or miltiple lines
lines = ['first line\n', 'secon line\n', 'third line\n']
with open('example.txt', 'a') as file:
    file.writelines(lines)
```

```python
### Binary Files
# writing to a binary file: 
data = b'\x00\x01\x02\x03\x04'
with open('example.bin', 'wb') as file:
    file.write(data)
```

```python
# writing to a binary file: 
with open('example.bin', 'rb') as file:
    content = file.read()
    print(content)
```
