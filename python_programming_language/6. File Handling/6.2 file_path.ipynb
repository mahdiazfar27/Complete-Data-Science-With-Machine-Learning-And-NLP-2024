### file paths

```python
import os
print(os.getcwd())
```

```python
## creat a new directory
new_directory = "package"
os.mkdir(new_directory)
print(f"Directory '{new_directory}' created")
```

```python
## listing all the files in a directory
items = os.listdir('.')
print(items)
```

```python
### joining path
# most importat feature os using os package.
dir_name = 'folder'
file_name = 'file.txt'
#relative path
relative_path = os.path.join(dir_name,file_name)
print(relative_path)
```

```python
dir_name = 'folder'
file_name = 'file.txt'
# full path
full_path = os.path.join(os.getcwd(),dir_name,file_name)
print(full_path)
```

```python
### if a file is present of not
path = "example.txt"
if os.path.exists(path):
    print(f"the path '{path}' exists")
else:
    print(f"the path '{path}' does not exist")
```

```python
### checking if a path is a file of directory
import os

path = 'example.txt'
if os.path.isfile(path):
    print(f"the path '{path}' is a file.")
elif os.path.isdir(path):
    print(f"the path '{path}' is a directory.")
else:
    print(f"the path '{path}' is neither a file nor a directory.")
```

```python
### getting the absolute path instead of using the join
relative_path = 'example.txt'
absolute_path = os.path.abspath(relative_path)
print(absolute_path)
```
