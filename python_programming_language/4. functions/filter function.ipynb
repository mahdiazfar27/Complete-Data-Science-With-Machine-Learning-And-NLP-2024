### filter() func in python

filter() is used to select elements from an iterable base on a condition (True/False)
it checks each elements from the iterable using the functin
keeps only those element where the condition is True

```python
### filter syntax
filter(function, iterable)
```

```python
### filter with def func
def even(num):
    if num%2==0:
        return True
```

```python
lst = [1,2,3,4,5,6,7,8]
even_lsts = filter(even,lst)
print(list(even_lsts))
```

```python
### filter with lambda func
even_lst = filter(lambda x: x%2==0, lst)
print(list(even_lst))
```

```python
### filter with lambda func and multiple conditions
even_greater_list = list(filter(lambda x: x%2==0 and x>4, lst))
print(even_greater_list)
```

```python
### filter() to check if the age is greater than 23 in dictionary
people=[
    {'name': "krish", 'age': 32},
    {'name': "farabe", 'age': 25},
    {'name': "shammi", 'age': 23},
    {'name': "pritom", 'age': 31}
]
```

```python
def is_age_greater(person):
    return person['age']>23

age_greater_23 = list(filter(is_age_greater, people))
print(age_greater_23)
```
