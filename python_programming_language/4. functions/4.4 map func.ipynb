### map() func

map() is used to apply a func to each item in an iterable(like  list, tuple etc).
its very powerful and heavily used.

```python
### syntax
map(function, iterable)
map(lambda_func, lst)
```

```python
num_lst = [1,2,3,4,5,6]
squre_lst = map(lambda x: x**2, num_lst)
print(list(squre_lst))
```

map() func returns a map object. we convert it into list

```python
### map multiple iterables
num1 = [1,3,5,7]
num2 = [2,4,6,8]

added_num = list(map(lambda x,y: x+y,num1,num2))
print(added_num)
```

```python
### map() to convert list of string to integers
str_num = ['1','2','3','4']
int_num = list(map(int, str_num))
print(int_num)
```

```python
### map() to convert lower words to upper
```

```python

def get_name(person):
    return person['name']

people = [
    {'name': 'krish', 'age': 32},
    {'name': "farabe", 'age': 25},
    {'name': "shammi", 'age': 23}
]
```

```python
names_lst = list(map(get_name,people))
print(names_lst)
```
