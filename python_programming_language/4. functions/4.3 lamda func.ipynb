### Lambda function

A lambda function is a small anonymous (unnamed) function used for short, one-line operations.
Key Points
No def keyword but use "lamda" keyword
No function name
there can be any num of arguments but only one expression
Usually used for simple, short tasks
Often used with functions like map(), filter(), sorted()

```python
### syntax
lambda arguments: expression
```

```python
def add(a,b):
    return (a+b)
print(add(2,5))
```

```python
add = lambda a,b,c: a+b+c #here add is a func-variable
type(add)
```

```python
print(add(2,5,1))
```

```python
if_even = lambda num: num%2==0
```

```python
print(if_even(9))
```

```python
## map() - this applies to all items in a list
numbers = [1,2,3,4,5,6]
def square(numbers):
    return numbers**2
for number in numbers:
    print(square(number))
```

```python
## map() - this applies to all items in a list
lst = [1,2,3,4,5,6,7,8,9]
list(map(lambda x: x**2,lst))
```
