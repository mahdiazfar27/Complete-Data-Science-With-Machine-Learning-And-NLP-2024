### importing modules and packages in python

python hierarchy:
Library     ### collection of packages e.g. scikit-learn,pandas ###
    package     ### folders of modules e.g. sklear.preprocessing ###
        module      ### single .py files e.g. math.py ###
            function/class      ### defined inside module e.g. sqrt() ###
                statements      ### actual lines of codes ###

```python
### importing package 
import math
math.sqrt(2)
```

```python
### importing funcion from package
from math import sqrt, pi
print(sqrt(25))
print(pi)
```

### here
library -> python standard library

function -> sqrt()
