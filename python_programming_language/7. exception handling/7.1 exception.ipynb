```python
### exception handling 
# try, except block
try:
    a=b
except:
    print("the variable has not been assigned")
```

```python
# exception throws a class and we can catch the exception using that class.
try:
    a=b
except NameError as ex:
    print(ex)
```

```python
try:
    result=1/2
    print(result)
    a=b
except ZeroDivisionError as ex:
    print(ex)
    print("please enter the dominator other than zero")
```

```python
### handling any exceptions by the parent Exception class
try:
    result=1/2
    print(result)
    a=b
except ZeroDivisionError as ex:
    print(ex)
    print("please enter the dominator other than zero")
except Exception as ex1: # parent  exception class should be handled at the very last except block.
    print(ex1)
    print("the parent Exception class caught the exception and handled it")
```

```python
try:
    num = int(input("enter a number"))
    result = 10/num
    print(result)
except ValueError:
    print("this is not a valid number")
except ZeroDivisionError:
    print("the dominator cant be zero")
except Exception as ex:
    print(ex)
```

```python
### try, except, else block
# we should have the print in else block.
# then either the except block or the else block will be executed
# meaning else will be exe only if there is no exception
try:
    num = int(input("enter a number"))
    result = 10/num
except ValueError:
    print("this is not a valid number")
except ZeroDivisionError:
    print("the dominator cant be zero")
except Exception as ex:
    print(ex)
else:
    print(f"the division is: {result}")
```

```python
### try,except,else, finally
# finally block always be executed weither exception arises or not
try:
    num = int(input("enter a number"))
    result = 10/num
except ValueError:
    print("this is not a valid number")
except ZeroDivisionError:
    print("the dominator cant be zero")
except Exception as ex:
    print(ex)
else:
    print(f"the division is: {result}")
finally:
    print("execution completed")
```

```python
### example file handling and exception handling

try:
    file = open('example.txt', 'r')
    content = file.read()
    print(content)
except FileNotFoundError:
    print("the file does not exist")
finally:
    if 'file' in locals() and not file.closed():
        file.close()
        print("file is closed")
```
