### Iterators (remember we use it a lot in c++ stl)

iterators are advanced python concepts that allow for efficient looping and memory management.
iterators provide a way to access elements of a collection sequentially without exposing the underlying structure.

```python
# for each loop
my_list = [1,2,3,4,5]
for i in my_list:
    print(i)

print(my_list)
type(my_list)
```

```python
### lopping using iterator
iterator = iter(my_list)
print(type(iterator))
```

```python
iterator
# its a list_iterator at this particular memory location.
# its a Lazy Loading Technique meaning it will not 
# give use any element unless we iterate through the memory location.
```

```python
### iterate through all the element
next(iterator)
```

```python
my_list = [1,2,3,4,5]
iterator = iter(my_list)
```

```python
try:
    print(next(iterator))
except StopIteration:
    print("there are no more elemets in the iterator")
```
