### Decorators

Decorators are a powerful and flexible feature in python that allows to modify the behavior of a function or class method.
They are commonly use to add functionality to functions or methods without modifying their actual code. 

Before learning decorators we need to learn
1. function copy
2. closures
3. decorators

```python
### function copy
def welcome():
    return "welcome to the decorators lecture"

welcome()
```

```python
wel = welcome #now wel() is a copy of welcome() func
wel
```

```python
print(wel())
del welcome # even after delete of the main or reference func the copy remains
print(wel())
```

```python
### closures - its basically a function inside another func.
# a closure is a func that remembers variables from its outer
# func even after the outer func has finished executing.
# basically by returning the inner func inside outer func we can create a closure
def outer(x):
    def inner(y):
        return x+y
    return inner
```

```python
f = outer(10) #--> f = return inner
print(f(5))   #--> f(5) = return inner(5) --> return (10+5)
```

```python
### closures functions
def main_welcome(str):
    msg = 'welcome' #its a local variable for the outer func
    def sub_welcome_method():
        print(str)
        print("welcome to the understanding of closures")
        print(msg) # for closure func they can actually acces the local variables of the outer func.
        print("please learn this topic properly")
    return sub_welcome_method()
```

```python
main_welcome("hello")
```

### Now its time for Decorators

```python
### Decorators

def main_welcome(func):
    
    def sub_welcome_method():
        print("welcome to the understanding of closures")
        func() 
        print("please learn this topic properly")
    return sub_welcome_method()
```

```python
def course_intoduction():
    print("this is an advance python course")
course_intoduction()
```

```python
main_welcome(course_intoduction)
```

now our chalange is without calling the main_welcome() can we get the :
"welcome to the understanding of closures
this is an advance python course
please learn this topic properly"

and to fulfill this desire we use Decorators

```python
# manual creation of decorator

@main_welcome   #this func is a decorator.
def course_intoduction(): #we are adding funtionallities without changing the actual codes of the decorator func.
    print("This is an advance python course")
```

```python
### exp2 decorator

#closure:
def my_decorator(func):
    def wrapper():
        print("something is happening before the func in called")
        func()
        print("something is happening after the func is called")
    return wrapper
```

```python
@my_decorator
def say_hello():
    print("hello")
```

```python
say_hello()
```

```python
### decorators with arguments
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):  # _ we use when we dont care about the use of another iterator variable.
                func(*args, **kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello():
    print("hello")
```

```python
say_hello()
```
