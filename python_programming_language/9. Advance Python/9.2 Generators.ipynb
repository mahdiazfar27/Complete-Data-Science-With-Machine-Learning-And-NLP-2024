### Generators

it is a subclass of iterators. Generators are a simpler way to create iterators.
They use the "yield" keyword to produce a series of values lazily, which means they generate values on the fly and do not store them in memory.

```python
def square(n):
    for i in range(3):
        return i**2
```

```python
print(square(3))
# here what happens that i gets first element of the range
# and does 0**2 and return 0 and the execution ends
# but instead of return we could use yield
# that would save the loops full executions values and provide
# a generator object by iterating which we can find the values
# basically yield returns  the full execution results to from
# where it had been called e.g. here square(3).
```

```python
### yield
def square(n):
    for i in range(3):
        yield i**2
```

```python
square(3)
```

```python
for i in square(3):
    print(i)
```

```python
a = square(3)
a
```

```python
next(a)
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
```

```python
gen = my_generator()
gen
```

```python
next(gen)
```

```python
for i in my_generator():
    print(i)
```

### practical example of generator
### reading large files:

because generator allow to process one line at a time without loading the entire file into the memory

```python
## reading large file

def read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line
```

```python
file_path = 'large_file.txt'
gen = read_large_file(file_path)
```

```python
next(gen)
```

```python
for line in read_large_file(file_path):
    print(line.strip())
```
