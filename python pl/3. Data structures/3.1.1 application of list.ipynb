### Real world examples using list in python

#### Ex1 manage a todo list

```python
to_do_list = ["buy gorceries", "clean the house", "pay the bills"]

#adding to task
to_do_list.append("schedule meeting")
to_do_list.append("go for a run")

#remove a completed task
to_do_list.remove("clean the house")

#checking if ta task is in the list
if "pay the bills" in to_do_list:
    print("hey!! dont forget to pay the bills")

print("ramaining tasks: ")
for task in to_do_list:
    print(f"--{task}")
```
