### Tuples

a tuple is a ordered collection like a list, but immutable, meaning its elements cannot be changed after creation and we use () for tuple.

so list = []
and tuple = ()

```python
### creating tuple
empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
## creating using the class

lst = list()
tpl = tuple()

print(type(lst))
print(type(tpl))
```

```python
### converting list to tuple and vice verca
numbers=tuple([1,2,3,4,5,6])
print(numbers)

print(list((1,2,3,4,5,6)))
```

```python
mxd_tuple = (1,2,"Farabe", 3.1416, True)
print(mxd_tuple)
```

```python
## accessing element
print(mxd_tuple[3])
print(mxd_tuple[-1])
print(mxd_tuple[1:3])
print(mxd_tuple[1:])
print(mxd_tuple[::2])

```

```python
### Tuples operations
concatinate_tuple = numbers + mxd_tuple
print(concatinate_tuple)
```

```python
mxd_tuple * 3
```

```python
# but after multiplication the tuple isnt changed
mxd_tuple
```

```python
### Immutable nature of tuples

lst = [1,2,"farabe", 5, 6]
lst[2] = "Krish"
print(lst)
```

```python
tpl = (1,2,"farabe",5,6)
tpl[2] = "krish"
print(tpl)
# so to change the element we need to change it to list firt then we can change the element then convert it to tuple again.
```

```python
### Packing and unpacking tuple
#packing

packed_tuple = 1,"hello",3.14
print(packed_tuple)
```

```python
#unpacking
a,b,c = packed_tuple
print(a)
print(b)
print(c)
```

```python
# unpacking with star(*)
print(numbers)
first, *middle, last = numbers
print(first)
print(middle)
print(last)
```

```python
### nested tist

lst = [[1,2,3,4],[6,7,8,9], [1,"hello",3.1416,"c"]]
print(lst[0])
print(lst[0][2])
print(lst[0][0:3])
```

```python
### nested tuple

nested_tpl = ((1,2,3,4),(6,7,8,9), (1,"hello",3.1416,"c"))

print(nested_tpl[1])
print(nested_tpl[0][2])
```

```python
# iterating over nested tuple
for sub_tpl in nested_tpl:
    for element in sub_tpl:
        print(element, end=" ")
    print()
```
