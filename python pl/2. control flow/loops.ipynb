## Loops(for, while, break, continue, pass,nested loop)

```python
range(5)
```

```python
for i in range(5):
    print(i)
```

```python
for i in range(1,8):
    print(i)
```

```python
#jumping 2 steps
for i in range(1,10,2):
    print(i)
```

```python
#jumping 1 steps in reverse
for i in range(10,1,-1):
    print(i)
```

```python
#jumping 2 steps in reverse
for i in range(10,1,-2):
    print(i)
```

```python
## Strings

str = "mahdi azfa"

for i in str:
    print(i)
```

```python
## While loop

count=0

while count<5:
    print(count)
    count+=1
```

```python
## loop control statement
#break

for i in range(10):
    if i==5:
        break
    print(i)
```

```python
# continue
for i in range(10):
    if i==5:
        continue
    print(i)
```

```python
## pass
# the pass statement is a null operation; it does nothing at all; at just stays for no reason

for i in range(5):
    if i==3:
        print("this is under pass ", i)
        pass
    print(i)
```

```python
## Nested loop
## f-strings; formated string printing where we use variable inside the curly brackets

for i in range(5):
    for j in range(2):
        print(f"i: {i} and j: {j}")

```

```python
## sum of first n natural numbers

n = int(input("enter the number of n"))
sum = 0
for i in range(1,n+1):
    sum+=i
print(f"sum of first {n} natural numbers is {sum}")
```

```python
## prime numbers between 1 to 100
import math

for i in range(2,101):
    is_prime = True
    for j in range(2, i):
        if i%j ==0:
            is_prime = False
            break
    if is_prime:
        print(i)
        
        
    
```
