```python
## real function
def even_or_odd(num):
    """this func find a num even or odd"""
    if num%2==0:
        print(f"the num {num} is even")
    else:
        print(f"the num {num} is odd")
    return
```

```python
## call this function
n = int(input("write a integer number"))
even_or_odd(n)
```

```python
## func with multiple parameter
def add(a,b):
    # sum = a+b
    return a+b

print(add(2,3))
```

```python
#default parameter
def greet(name):
    print(f"hello {name} welcome to python coding!")

greet()
```

```python
### variable length arguments
## 1) positional arguments ## when the arguments are normal values

#def print_func(num1, num2, num3, num4):
#instead of using this multiple arguments we can use "*argument_name"

def print_func(*args):
    for i in args:
        print(i, end=" ")
        
```

```python
print_func(1,2,3,6.6666,"farabe", True)
```

```python
## 2) keywords arguments ## when the arguments are key-value pair

def print_key_val(**kwargs):
    for key,value in kwargs.items():
        print(f"{key} : {value}")
```

```python
print_key_val(name="farabe", age=25, height=5.7)
```

```python
## now combining two variable length arguments

def print_everything(*args, **kwargs):
    print("positional arguments:")
    for val in args:
        print(f"{val}", end=" ")
    print()
    print("keyword arguments:")
    for key, values in kwargs.items():
        print(f"{key}: {values}")
```

```python
print_everything(1,2,4,5.555,True,"mahdi", name="farabe", age=25, height=5.7)
```
