### exp1: temp converter

```python
def convert_temp(temp,unit):
    """Docstring"""
    if unit=="c":
        return temp * 9/5 + 32
    elif unit=="f":
        return (temp-32) * 5/9
    else:
        return None
```

```python
print(convert_temp(25,"c"))
print(convert_temp(77,"f"))
```

### exp2 password strength checker

```python
def is_strong_pass(password):
    if len(password) < 8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char in '!@#$%^&*()_+' for char in password):
        return False
    return True
    
```

```python
print(is_strong_pass("12345"))
print(is_strong_pass("Strongp@ss26"))
```

### exp3 total cost of a shopping cart

```python
def calculate_cost(cart):
    total = 0
    for item in cart:
        total += (item["unit_price"]*item["quantity"])
    return total
```

```python
cart=[
    {"name":"banana", "unit_price": 5, "quantity": 12},
    {"name":"orrange", "unit_price": 15, "quantity": 10},
    {"name":"guava", "unit_price": 8, "quantity": 5},
    {"name":"apple", "unit_price": 5, "quantity": 4},
    {"name":"strawberry", "unit_price": 50, "quantity": 7},
]

print(f"total: {calculate_cost(cart)} tk")
```

### exp4 check if a string is palindrom

```python
def is_palindrom(s):
    s = s.lower().replace(" ","")
    return s == s[::-1]
```

```python
s = input("write a string")
if(is_palindrom(s)):
    print(s, ": palindrom")
else:
    print(s, ": not palindrom")
```

### exp5 calculate the factorial of a number using recursion

```python
def factorial(num):
    if num == 1:
        return 1
    else:
        return num * factorial(num-1)
```

```python
n = 5
print(factorial(n))
```

### exp6 a func to read a file and count the frequency of each word

```python
def count_word_freq(file_path):
    # i will use a dict to store in key value for the word conunt.
    word_count = {}
    # to open a file we use open func
    with open(file_path,'r') as file:
        for line in file:
            words = line.split()
            for word in words:
                word = word.lower().strip('.,!/?;:"\'')
                word_count[word] = word_count.get(word,0)+1
    return word_count
```

```python
file_path = "file_to_read.txt" # as both files are in the same directory
print(count_word_freq(file_path))
```

### exp7 validate email address

```python
import re
# re is a Regular Expression Module used for pattern matching
def is_email_valid(email):
    pattern = r'^[a-zA-Z0-9._+-]+@[a-zA-Z0-9-]+\.[a-zA-Z0-9-.]+$'
    return re.match(pattern,email) is not None
```

```python
print(is_email_valid("mahdiazfar@gmail.com"))
```
